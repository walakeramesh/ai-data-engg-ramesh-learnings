# List

Its a built in datatype used to store multiple items in a single variable

Key Feature of List
* Ordered = Items have fixed position
* Mutable = You can change, add or remove
* Allow Duplicates = Same value can appear multiple times 
* Can store different datatype = number, string, even list and etc

```python
myList=[1,2,3,4,5,6,7]
print(myList)
```

```python
myVariable=["AI","BI","MLOPS",1,2,3,4,5,6,[23,32,33],False,True,None]
print(myVariable)
```

```python
type(myVariable)
```

```python
#append
myVariable.append(12345)
print(myVariable)
```
