# Set

Set in Python as a built-in data type used to store a collection of unique, unordered elements.
#Key FeaturesUnordered: 
* Items do not have a fixed position, meaning they cannot be accessed via an index.
* No Duplicates: Each value in a set must be unique; it appears only once.
* Mutable: You can add or remove items after the set has been created

```python
mySet=set() # empty set
print(mySet)
```

```python
mySet={1,2,3,4,5,6,7}
print(mySet)
```

```python
mySet={"DE","ML","AI","Gen-Ai","ML"}
print(mySet) # result order changes and no duplicates
```

## Operation on Set

```python
mySet.add("math") # Addig new value to set
print(mySet)
```

```python
mySet[3]  # it doesn't support for indexing
```

```python
mySet1={"df","raw","curated","silver","python"}
mySet2={"df","raw","curated","silver","python","scala"}
```

```python
print(mySet2.difference(mySet1)) 
```

```python
print(mySet2.intersection(mySet1)) 
```

# Dictionaries 

A built-in data type used to store data in key-value pairs.
# Key FeaturesKey-Value Structure: 
* Each item consists of a key and its corresponding value.
* Mutable: You can change, remove, and add items after the dictionary is created.
* Unique Keys: Keys must be unique; no duplicates are allowed.
* Ordered: (Python 3.7+) Items keep their insertion order.

```python
myDict={"name":"Ramesh","age":30,"city":"belgaum"}
print(myDict)
```

```python
for key in myDict: # To print only keys
    print(key)
```

```python
for key in myDict.values(): # To print values
    print(key)
```

```python
for key in myDict.items(): # print both key values
    print(key)
```

```python
myDict["country"]="India" # add new key value pair
print(myDict)
```

```python
myDict["age"]=29 # update any key from dictionary
print(myDict)
```

### Nested Dictionary

```python
students = {
    "Alice": {
        "age": 20,
        "grade": "A",
        "subjects": ["Math", "Science", "English"]
    }
}
print(students)
type(students)
```

```python
students["Alice"]["subjects"]
```
